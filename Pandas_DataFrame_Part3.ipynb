### 11)Compare the number of survived passengers on the basis of their sex:

```python
pd.crosstab( titanic.sex, titanic.survived )  #From survived, there were 339 female and 161 male.
```

```python
pd.crosstab( titanic.sex, titanic.survived, normalize = "index" ) #normilizing on the basis of index.
```

```python
pd.crosstab( titanic.sex, titanic.survived, normalize = "columns" ) #normalizing in the basis of columns.
```

```python
pd.crosstab( titanic.sex, titanic.survived, normalize = "all" ) #normalizing on the basis of whole data.
```

### 14)The number of children below 6, who are survived:

```python
titanic[ titanic.age<=6 ][["survived","age"]]
```

```python
pd.crosstab( titanic.age<=6, titanic.survived)
```

```python
titanic[ titanic.age <= 6 ]["survived"].value_counts( )
```

```python
titanic[ titanic.age <= 6 ]["survived"].value_counts(normalize =True)*100  #nearly 65 percent of children have been survived.
```

- The value_counts() function returns a Series that contain counts of unique values. It returns an object that will be in descending order so that its first element will be the most frequently-occurred element. By default, it excludes NA values.

### 16) unique data in a column:
##### print the data about sex, age and Pclass of whom survived.

```python
titanic[ ( titanic.survived == 1 ) & ( titanic.age <= 6 ) ][['age','sex','pclass']]
```

##### The null data 

```python
titanic[ titanic.age.isnull() ][['age', 'survived', 'sex', 'pclass']]
```

```python
titanic[ -titanic.age.isnull() ][['age',
                                          'survived',
                                          'sex',
                                          'pclass']][0:5]    #the not null data printing
```

### 17)How to merge and group different columns

```python
pclass_age_mean_df = titanic.groupby( 'pclass' )['age'].mean()
pclass_age_mean_df
```

```python
pclass_gender_age_mean_df = titanic.groupby( ['pclass', 'sex'] )['fare'].mean().reset_index()
pclass_gender_age_mean_df
```
